```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv("Supermarket.csv")
df
```

```python
# Convert Products column to a list of lists
df['Products'] = df['Products'].apply(lambda x: x.split(','))

# Display the preprocessed dataset
print(df.head())
```

```python
# Convert the list of lists into a one-hot encoded DataFrame
te = TransactionEncoder()
onehot = te.fit(df['Products']).transform(df['Products'])
onehot_df = pd.DataFrame(onehot, columns=te.columns_)

# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(onehot_df, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)

# Display the generated rules
print(rules)
```

```python
# Analyze the generated rules
# Filter rules with high confidence
high_confidence_rules = rules[rules['confidence'] > 0.7]

# Sort rules by lift
sorted_rules = high_confidence_rules.sort_values(by='lift', ascending=False)

# Display the sorted rules
print(sorted_rules)
```
